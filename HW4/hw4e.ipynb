```python
''''''Перечень команд
     по команде “p” можно узнать владельца документа по его номеру
     по команде “s” можно по номеру документа узнать на какой полке он хранится
     по команде “l” можно увидеть полную информацию по всем документам
     по команде “ds” может удалить существующую полку из данных (только если она пустая)
     по команде “as” может добавить новую полку''''''   
```

```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def multi_pass():
    docnum = input('Введите номер документа:')
    for document in documents:
        if docnum ==  document['number']:
            print('Владелец документа: ', document['name'])
            break
    else:        
        print('Документ не найден в базе')
        
def shelf_num(dir, value):
        for shelfnum, numbers in directories.items():
            if value in numbers:
                return shelfnum
            
def shelf_doc():
        docnum = input('Введите номер документа:')
        res = (shelf_num(directories, docnum))
        if res is None:
            print('Документ не найден в базе')
        else:
            print('Документ хранится на полке:​ ', res)
            
    
def all_docs():
    for docs in documents:
        print('№: ', docs['number'], 'тип:', docs['type'], \
        'владелец:', docs['name'], 'полка хранения: ', shelf_num(directories, docs['number']))
        
def new_shelf():
    shelfnum = input('Введите номер полки: ')
    if shelfnum in directories.keys():
            print('Такая полка уже существует. Текущий перечень полок:', (', '.join(directories.keys())))
    else:
        directories.setdefault(shelfnum, [])
        print('Полка добавлена. Текущий перечень полок:', (', '.join(directories.keys())))
        
def dlt_shelf():
    shelfnum = input('Введите номер полки: ')
    for k, v in list(directories.items()):
        if shelfnum is k:
            if not v:
                res = directories.pop(k, None)
                print('Полка удалена. Текущий перечень полок:', (', '.join(directories.keys())))
                break
            else:
                print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:'\
                , (', '.join(directories.keys())))
                break
    else:
        print('Такой полки не существует. Текущий перечень полок:', (', '.join(directories.keys())))
    
def main():
    while True:
        user_input = input('Введите команду: ')
        if user_input == 'p':
            multi_pass()
        elif user_input == 's':
            shelf_doc()
        elif user_input == 'l':
            all_docs()
        elif user_input == 'as':
            new_shelf()
        elif user_input == 'ds':
            dlt_shelf()
        elif user_input == 'q':
            print('Завершение программы')
            break
        else:
            print('Неадекватная команда, введите корректную команду или завершите работу командой q ')
            
main()           
```
