```python
import pandas as pd
r = pd.read_csv('ratings.csv')
def sorting(row):
    if row <= 2:
        return 'Низкий'
    elif row <= 4:
        return 'Средний'
    elif row == 4.5 or 5:
         return 'Высокий'
    else:
        return 'Неопределённый'
r['class'] = r['rating'].apply(sorting)
r.head(5)
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

import pandas as pd
kw = pd.read_csv('keywords.csv')

new_dic = {}

for k,v in geo_data.items():
    for x in v:
        new_dic.setdefault(x,[]).append(k)


def geoclass(row):
    for k, v in new_dic.items():
        if k in row: 
            return v[0]
    else:
        return 'undefined'
kw['region'] = kw['keyword'].apply(geoclass)
kw.head(35)
```

```python
import pandas as pd
movies = pd.read_csv('movies.csv')
r = pd.read_csv('ratings.csv')
movies['year'] = movies['title'].str.extract('.*\((.*)\).*',expand=True)
movies['year'] = pd.to_numeric(movies['year'], errors = 'coerce')       
movies = movies.dropna(subset=['year'])
movies['year'] = movies['year'].astype(int)
movies.query('year >= 1950 & year <= 2011', inplace = True)
movrat = r[['movieId', 'rating', 'userId']].merge(movies[['movieId', 'title', 'year']],\
on ='movieId', how = 'right')
movrat.groupby(['year'])['rating'].mean().sort_values(0, False).head(15)
```
